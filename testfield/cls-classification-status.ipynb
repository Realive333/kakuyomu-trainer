```python
import json
import pandas as pd
from statistics import mean
```

```python
TARGET = [1, 2, 3, 4, 5, 20, 39, 40, 42, 69, 70, 71, 73, 74, 75, 77, 79, 80, 81, 83, 84, 87, 90, 96, 120, 121, 122, 126, 128, 199, 200, 203, 204, 214, 259, 260, 281, 284, 291]
```

```python
res = []
for target in TARGET:
    path = f"../tsv/first-match-scatter/{target}/cls-result.json"
    with open(path, 'r') as f:
        results = json.load(f)
    json_result = []
    logc = 0
    avgc = 0
    gndc = 0
    for r in results:
        label = r['label']
        if label == 1:
            index = 0
        else:
            index = 1
        logits = r['logits_softmax'][0][index]
        #logits = mean([i[index] for i in r['logits_softmax']])
        avg = r['avg_softmax'][index]
        gnd = r['grand_softmax'][index]
        json_result.append({
            'label': label,
            'logits': logits>0.5,
            'avg': avg>0.5,
            'gnd': gnd>0.5
        })
        if logits > 0.5:
            logc += 1
        if avg > 0.5:
            avgc += 1
        if gnd > 0.5:
            gndc += 1
    res.append([logc/len(results), avgc/len(results), gndc/len(results)])
```

```python
pd.DataFrame(res)
```
