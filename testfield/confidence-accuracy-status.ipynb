```python
import json
import pandas as pd
```

```python
with open('./label-target.json') as f:
    label = json.load(f)
```

```python
TARGET = [1, 2, 3, 4, 5, 20, 39, 40, 42, 69, 70, 71, 73, 74, 75, 77, 79, 80, 81, 83, 84, 87, 90, 96, 120, 121, 122, 126, 128, 199, 200, 203, 204, 214, 259, 260, 281, 284, 291]
MODEL = 'LinearRegression'
result = []
for target in TARGET:
    path = f'../savepoint/regression/{MODEL}/{target}'
    with open(f'{path}/test-accuracy.json', 'r') as f:
        accuracy = json.load(f)['accuracy']
    result.append({'target': label[str(target)], 'acc': accuracy})
```

```python
pd.DataFrame(result)
```
