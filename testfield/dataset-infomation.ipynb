```python
import os
import csv
import sys
import workFactory as wf
from tqdm import tqdm
csv.field_size_limit(sys.maxsize)
```

```python
file_dir = '../tsv/full'
dirs = []

for filename in os.listdir(file_dir):
    f = os.path.join(file_dir, filename)
    dirs.append(f)
```

```python
tsv = wf.load_tsv_dataset('../tsv/full/42/test.tsv')
print(len(tsv))
print(sum(len(i['content']) for i in tsv))
sum(len(i['content']) for i in tsv)/len(tsv)
```

```python
results = []
for folder in tqdm(dirs):
    train = wf.load_tsv_dataset(f'{folder}/train.tsv')
    tr_avg = sum(len(i['content']) for i in train)/len(train)
    
    tests = wf.load_tsv_dataset(f'{folder}/test.tsv')
    t_avg = sum(len(i['content']) for i in tests)/len(tests)
    
    dev = wf.load_tsv_dataset(f'{folder}/dev.tsv')
    d_avg = sum(len(i['content']) for i in dev)/len(dev)
    
    #print(f'{folder}\n\t{tr_avg}\n\t{t_avg}\n\t{d_avg}')
    results.append({'id': folder, 'train': tr_avg, 'test':t_avg, 'dev':d_avg})
```

```python
import pandas as pd
df = pd.DataFrame(results)
```
