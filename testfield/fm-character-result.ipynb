```python
import json
from tqdm import tqdm
```

```python
def isSuccess_majority(work):
    scores = work['confidence_scores'].replace('[', '').replace(']', '').split(', ')
    success = 0
    failed = 0
    for score in scores:
        if float(score) > 0.5:
            success += 1
        else:
            failed += 1
    if success > failed:
        return True
    else:
        return False
```

```python
def isSuccess_average(work):
    scores = work['confidence_scores'].replace('[', '').replace(']', '').split(', ')
    total = 0
    for score in scores:
        total += float(score)
    average = total/len(scores)
    if average > 0.5:
        return True
    else:
        return False
```

```python
def getAverage(works, method):
    success = 0
    failed = 0
    for work in works:
        if method(work):
            success += 1
        else:
            failed += 1
    return (success/(success+failed))
```

```python
results = []

targets = [1, 2, 3, 4, 5, 20, 39, 40, 42, 69, 70, 71, 73, 74, 75, 77, 79, 80, 81, 83, 84, 87, 90, 96, 120, 121, 122, 126, 128, 199, 200, 203, 204, 214, 259, 260, 281, 284, 291]
for target in tqdm(targets):
    path = f'../tsv/first-match-character/{target}/fm-confidence.json'
    with open(path, 'r', encoding='utf-8') as f:
        works = json.load(f)
        
    maj = getAverage(works, isSuccess_majority)
    avg = getAverage(works, isSuccess_average)

    results.append({
        'label': target,
        'accuracy_majority': maj,
        'accuracy_average': avg
    })
    
with open('../tsv/first-match-character/fm-result.json', 'w+') as f:
    json.dump(results, f)
```
