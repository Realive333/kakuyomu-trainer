```python
import re
import csv
import sys
import MeCab

import workFactory

csv.field_size_limit(sys.maxsize)
```

```python
TARGET = 42
PATH = f'./tsv/front-512/{TARGET}'
```

```python
def analysis(work):
    surface = []
    cnt = work['content']
    sep = re.split("。|「|」|？|！", cnt)
    mecab = MeCab.Tagger("-Owakati")
    for sent in sep:
        phrase = mecab.parse(sent).replace('\n', '')
        surface.append(phrase)
    surface = [x for x in surface if x]
    return '\n'.join(surface)
```

```python
def getMorphemeByTSV(path, target):
    workList = []
    works = workFactory.load_tsv_dataset(f'{path}/{target}.tsv')
    for work in works[:5]:    ##
        morp = analysis(work)
        if work['label'] == 0:
            label = '01'
        elif work['label'] == 1:
            label = '10'
        workList.append({'label': label, 'content': morp})
    return workList
```

```python
getMorphemeByTSV(PATH, 'dev')
```
