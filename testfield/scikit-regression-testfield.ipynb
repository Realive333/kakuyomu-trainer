```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_boston

boston = load_boston()

df = pd.DataFrame(boston.data, columns = boston.feature_names)
df['MEDV'] = boston.target

df.head()
```

```python
import seaborn as sns
sns.pairplot(df, size=1.0)
```

```python
sns.regplot('RM', 'MEDV', data=df)
```

```python
df.corr()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 0:13], df.iloc[:, 13], test_size=0.2, random_state=1)

sc = StandardScaler()
sc.fit(x_train) #学習用データで標準化
x_train_std = sc.transform(x_train)
x_test_std = sc.transform(x_test)
```

```python
print(x_train[:5])
print(x_train_std[:5])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error

lr = LinearRegression()
lr.fit(x_train_std, y_train)

pred_lr = lr.predict(x_test_std)

r2_lr = r2_score(y_test, pred_lr)
mae_lr = mean_absolute_error(y_test, pred_lr)
print(pred_lr)

print("R2 : %.3f" % r2_lr)
print("MAE: %.3f" % mae_lr)

print("Coef = ", lr.coef_)
print("Intercept = ", lr.intercept_)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.xlabel("pred_lr")
plt.ylabel("y_test")
plt.scatter(pred_lr, y_test)
plt.show()
```

```python
from sklearn.linear_model import Ridge

ridge = Ridge(alpha=10)
ridge.fit(x_train_std, y_train)

pred_ridge = ridge.predict(x_test_std)

r2_ridge = r2_score(y_test, pred_ridge)
mae_ridge = mean_absolute_error(y_test, pred_ridge)

print("R2 : %.3f" % r2_ridge)
print("MAE : %.3f" % mae_ridge)

print("Coef = ", ridge.coef_)
```

```python
plt.xlabel("pred_ridge")
plt.ylabel("y_test")
plt.scatter(pred_ridge, y_test)

plt.show()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.05)
lasso.fit(x_train_std, y_train)

pred_lasso = lasso.predict(x_test_std)

r2_lasso = r2_score(y_test, pred_lasso)
mae_lasso = mean_absolute_error(y_test, pred_lasso)

print("R2 : %.3f" % r2_lasso)
print("MAE : %.3f" % mae_lasso)

print("Coef = ", lasso.coef_)
```

```python
plt.xlabel("pred_lasso")
plt.ylabel("y_test")
plt.scatter(pred_lasso, y_test)

plt.show()
```

```python
import json

nkPath = f'../tsv/nearest-k-candidates-offset-faster/n-100/o-256/{TARGET}/confidence/top-10'
with open(f'{nkPath}/test_result.json', 'r') as f:
    testWorks = json.load(f)['works']
```

```python
from tqdm import tqdm

def getRegAccByWorks(works):
    success = []
    failed = []
    for work in tqdm(works):
        contents = work['contents']
        confidences = work['confidence_scores']
        confidences = confidences.replace('[', '').replace(']', '').replace(' ', '').split(',')
        reg_confidence = []
        for content in contents:
            paragraph = content['content'].replace('[SEP]', '')
            reg_pred = classifier.predict([paragraph])
            reg_confidence.append(reg_pred[0])
        reg_max = max(reg_confidence)
        reg_maxidx = reg_confidence.index(reg_max)
        pred_conf = confidences[reg_maxidx]
        if float(pred_conf) > 0.5:
            success.append(pred_conf)
        else:
            failed.append(pred_conf)
    print(success)
    print(failed)
    print(f'acc: {len(success)/(len(success)+len(failed)):.4f}')
    return len(success)/(len(success)+len(failed))
getRegAccByWorks(testWorks)
```

```python
print(f'acc: {success/(success+failed)}')
```
